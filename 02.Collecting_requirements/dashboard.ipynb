2. Понимание задачи
	Фирма занимается продажей товаров.
Я работаю в отделе HR аналитики и одна из моих задач – эффективное распределение премий среди сотрудников.
Я руковожу финансовым отделом и занимаюсь распределением премиального бюджета.
 Чтобы его распределить, мне нужно отобрать сотрудников для получения соответствующих премий.
 Для анализа по каждой должности есть свой HR-менеджер, который сравнивает сотрудников друг с другом внутри одной должности.
По рейтингу я определяю уровень, какой % от зарплаты будет выдан в качестве дополнительной премии.
Для принятия решения о премии также во внимание берется самооценка сотрудником удовлетворенности работой, стаж работы и уровень зарплаты.

5. Вопросы и бизнес-решения
	Финальные значения процентов определяет HR-менеджер индивидуально для каждого сотрудника
	Для точного определения процентов внутри диапазона используется анализ количества закрытых проектов за рабочий период.
	Какова  удовлетворенность  работой?
	Какой уровень удовлетворенности внутри компании?
	Как  стаж вне компании влияет на рейтинг?
	У каких работников больше шансов получить премию в следствии закрытых проектов?
	Как влияет обучение персонала на закрытие проектов?
	Какой % от зарплаты будет выдан в качестве дополнительной премии?
6. Данные
	У нас есть данные про каждого сотрудника в компании, информация о его рабочей позиции, социально-демографические и профессиональные характеристики.
	Эксель-Файлы:
	'satisfaction' – вариант для задания с оценкой удовлетворенности сотрудников; 
'monetary' – вариант для распределения премий.
	
	Основная метрика для оценки сотрудников – performance rating (далее рейтинг), он может принимать значения от 1 до 4. 1 — плохая работа, 2 — в рамках ожиданий, 3 — выше ожиданий, 4 — превосходные результаты.
 По рейтингу вы определяете уровень, какой % от зарплаты будет выдан в качестве дополнительной премии.
Для точного определения процентов внутри диапазона используется анализ количества закрытых проектов за рабочий период (projects closed).
